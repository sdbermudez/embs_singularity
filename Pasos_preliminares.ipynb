## RUTA A — Configuración local con Git (recomendada)

Usa esta ruta si ya tienes instalado:
- Python 3.8 o superior
- Git
- Un IDE como VS Code, PyCharm o Jupyter Lab

> **¿No sabes si tienes Git?** Abre una terminal y escribe `git --version`. Si ves un número de versión, lo tienes instalado. Si no, ve a la **Ruta B**.

### Paso A-1 — Clonar el repositorio

Abre **Git Bash** (Windows) o la **Terminal** y navega a la carpeta donde quieres guardar el proyecto. Luego ejecuta el siguiente comando. En windows se puede hacer algo como esto para navegar a la carpeta destino: 

````python
! cd documents/...
````

Este comando descarga una copia completa del repositorio de GitHub a tu computadora, creando una carpeta llamada `embs_singularity`.

````python
! git clone https://github.com/sdbermudez/embs_singularity.git
````

### Paso A-2 — Entrar a la carpeta del proyecto

Muévete dentro de la carpeta que acaba de crearse. Reemplaza `...` con la ruta completa donde hiciste el clone (por ejemplo, en Windows sería algo como `C:/Users/TuNombre/Documents`).

> También puedes abrir directamente la carpeta `embs_singularity` desde tu IDE arrastrándola a la ventana del programa.

````python
! cd .../embs_singularity
````

### Paso A-3 — Crear un entorno virtual

Abre **Windows PowerShell** o **CMD** (no Git Bash para este paso) y navega hasta la carpeta del proyecto. Luego ejecuta:

Un **entorno virtual** es un espacio aislado de Python donde instalaremos las librerías del proyecto sin afectar el resto de tu computadora. Se creará una carpeta llamada `sgem` dentro del proyecto.

````python
! python -m venv sgem
````

### Paso A-4 — Activar el entorno virtual

Debes activar el entorno virtual que acabas de crear. Esto le indica a Python que use ese espacio aislado.

Una vez activado, verás que el nombre `(sgem)` aparece al inicio de tu terminal, indicando que el entorno está activo.

````python
! sgem\Scripts\activate
````

### Paso A-5 — Instalar las dependencias

Ahora instala todas las librerías que el proyecto necesita. El archivo `requirements.txt` ya incluye la lista completa de paquetes con sus versiones exactas.

Este proceso puede tomar **2 a 3 minutos** dependiendo de tu conexión a internet.

````python
! pip install -r requirements.txt
````

### Paso A-6 — Seleccionar el kernel en tu IDE

Para ejecutar el notebook `CardioAI.ipynb` con el entorno que acabas de crear, debes decirle a tu IDE que use `sgem` como kernel de Python.

**En VS Code:**
1. Abre el archivo `CardioAI.ipynb`
2. Haz clic en el botón **"Select Kernel"** (esquina superior derecha)
3. Selecciona **"Python Environments..."**
4. Elige el entorno que contiene `sgem`

**En Jupyter Lab / Notebook clásico:**
1. Ve al menú **Kernel → Change Kernel**
2. Selecciona **`sgem`**

> ¡Listo! Ahora puedes abrir `CardioAI.ipynb` y ejecutar todas las celdas.

## Solución de problemas comunes

| Problema | Causa probable | Solución |
|----------|---------------|----------|
| `python` no se reconoce como comando | Python no está en el PATH | Reinstala Python marcando "Add to PATH" |
| Error al activar el entorno en PowerShell | Política de ejecución de Windows | Ejecuta `Set-ExecutionPolicy RemoteSigned` como administrador |
| `FileNotFoundError: heart_disease_uci.csv` | El directorio de trabajo no es correcto | Asegúrate de estar dentro de la carpeta `embs_singularity` |
| En Colab: `Runtime disconnected` | Inactividad o sesión expirada | Vuelve a ejecutar desde el **Paso C-2** |
| Kernel muerto o reiniciado en Colab | Memoria insuficiente | Ve a **Runtime → Change runtime type → T4 GPU** |
